```python
print("dtype:", x2.dtype) #regresa el tipo de dato
```

```python
print("itemsize:", x3.itemsize, "bytes") #regresa el tamaño en bytes
print("nbytes:", x3.nbytes, "bytes") #son los bites del elemento x el numero
```

#####Estadistica Basica

Media (Mean)

```python
data = np.array([1, 2, 3, 4, 5])
media = np.mean(data)
print(media)
```

Promedio (Average) y Promedio Ponderado

```python
data = np.array([1, 2, 3, 4, 5])
weights = np.array([1, 2, 3, 4, 5])
promedio = np.average(data)
promedio_ponderado = np.average(data, weights=weights)
```

```python
print(promedio)
print(promedio_ponderado)
```

Mediana

```python
mediana = np.median(data)
print(mediana)
```

Varianza y Desviación Estándar


```python
varianza = np.var(data)
desviacion = np.std(data)
print(varianza)
print(desviacion)
```

Generación de Datos Aleatorios

```python
muestra = np.random.normal(0, 1, 1000) #genera numeros aleatorios con media 0, desviacion estandar 1 y 1000 datos
print(np.mean(muestra), np.std(muestra))
```

Correlación

```python
corr = np.corrcoef([1,2,3], [1,5,7])
print(corr)
```
