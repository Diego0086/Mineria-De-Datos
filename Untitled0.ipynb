##### Mineria de datos

Mineria de datos: Es el uso del machine learning y el análisis estadístico para descubrir patrones y otra información valiosa de grandes conjuntos de datos. El proceso de minería de datos implica varios pasos, desde la recopilación de datos hasta la visualización para extraer información valiosa de grandes conjuntos de datos.
