#####Estadistica Basica

Media (Mean)

```python
data = np.array([1, 2, 3, 4, 5])
media = np.mean(data)
print(media)
```

Promedio (Average) y Promedio Ponderado

```python
data = np.array([1, 2, 3, 4, 5])
weights = np.array([1, 2, 3, 4, 5])
promedio = np.average(data)
promedio_ponderado = np.average(data, weights=weights)
```

```python
print(promedio)
print(promedio_ponderado)
```

Mediana

```python
mediana = np.median(data)
print(mediana)
```

Varianza y Desviación Estándar


```python
varianza = np.var(data)
desviacion = np.std(data)
print(varianza)
print(desviacion)
```

Generación de Datos Aleatorios

```python
muestra = np.random.normal(0, 1, 1000) #genera numeros aleatorios con media 0, desviacion estandar 1 y 1000 datos
print(np.mean(muestra), np.std(muestra))
```

```python
# Variables
#df es dataframe que traera todos los datos y los convertira en numeros
df['edad'] = pd.to_numeric(df['¿Cuántos años cumplidos tienes?'], errors='coerce')
#asignacion de si sera true y no tendra falso
map_sn = {'Si': True, 'No': False}
#genera columna nueva del dataframe de prog_python y lo mapea
df['prog_python'] = df['¿Has realizado programas utilizando Python?'].map(map_sn)
df['usa_colab'] = df['¿Has utilizado Google Colab?'].map(map_sn)
df['cuenta_github'] = df['¿Tienes cuenta de GitHub?'].map(map_sn)
```

```python
# A NumPy
edad = df['edad'].to_numpy(dtype=float) #convierte el tipo de dato a flotante
prog_py = df['prog_python'].to_numpy(dtype=bool) #lo convierte a numpy de tipo booleano
usa_colab = df['usa_colab'].to_numpy(dtype=bool) #trae la columna y lo convierte a numpy de tipo booleano
cuenta_gh = df['cuenta_github'].to_numpy(dtype=bool)
```

```python
# Descriptivos de edad (NumPy) / otro diccionario
summary_age = {
    'n': int(edad.size), #total de los elementos con size en el arreglo de edad
    'min': float(np.min(edad)), #min traera la menor edad
    'p25': float(np.percentile(edad, 25)), #percentil traera el % de los elementos en este caso el 25%
    'mediana': float(np.median(edad)), #da la mediana de la edad
    'media': float(np.mean(edad)), #da el promedio de la edad
    'p75': float(np.percentile(edad, 75)), #da el 75% de la edad
    'max': float(np.max(edad)),
    'rango(ptp)': float(np.ptp(edad)), #da el rango del pico minimo al maximo
    'var_muestral': float(np.var(edad, ddof=1)), #varianza con rango de libertad de 1
    'std_muestral': float(np.std(edad, ddof=1)), #desviacion estandar con libertad de 1
    'cv': float(np.std(edad, ddof=1)/np.mean(edad)), #coficiente de varianza
}
```

```python
# Frecuencias (NumPy)
freq_python = {'Si': int(np.count_nonzero(prog_py)), 'No': int(edad.size - np.count_nonzero(prog_py))}
freq_colab  = {'Si': int(np.count_nonzero(usa_colab)), 'No': int(edad.size - np.count_nonzero(usa_colab))}
freq_gh     = {'Si': int(np.count_nonzero(cuenta_gh)), 'No': int(edad.size - np.count_nonzero(cuenta_gh))}
```

```python
# Tablas cruzadas (NumPy)
def tabla2x2(a, b): #funcion de tabla 2x2
    t = np.zeros((2,2), dtype=int) #np.zeros es un arreglo o matriz de 2 renglones y 2 coloumnas llenas de ceros en tipo entero
    t[1,1] = np.count_nonzero(a & b) #asigna valores en a y b
    t[1,0] = np.count_nonzero(a & ~b)
    t[0,1] = np.count_nonzero(~a & b)
    t[0,0] = np.count_nonzero(~a & ~b)
    return t

#realiza las multiplicaciones booleanas
tab_py_colab = tabla2x2(prog_py, usa_colab)
tab_py_gh    = tabla2x2(prog_py, cuenta_gh)
tab_colab_gh = tabla2x2(usa_colab, cuenta_gh)
```
